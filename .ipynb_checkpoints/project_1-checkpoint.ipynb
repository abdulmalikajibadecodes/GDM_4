#### Plot with 0 degrees celcius included 

```python
plt.figure(figsize=(10, 8))
plt.scatter(x=well_data.wlbTotalDepth, y=well_data.wlbBottomHoleTemperature, 
            color='blue', edgecolor='k', alpha=0.7)

plt.title('Well Total Depth vs Bottom Hole Temperature', fontsize=16)
plt.xlabel('Total Depth (ft)', fontsize=14)
plt.ylabel('Bottom Hole Temperature (°F)', fontsize=14)

plt.show()
```

#### 2. Make a scatter Plot of the drilling days (vertical axis) according to the total depth (horizontal axis); use query function to keep only 0 < wlbDrillingDays < 800;

```python
mask = (well_data['wlbDrillingDays'] > 0) & (well_data['wlbDrillingDays']< 800)

well_data_0_to_800dys = well_data[mask]

plt.scatter(well_data_0_to_800dys.wlbTotalDepth, well_data_0_to_800dys.wlbDrillingDays)
```

```python
# mask = (well_data['wlbDrillingDays'] > 0) & (well_data['wlbDrillingDays'] < 800)
# well_data_0_to_800dys = well_data[mask]

well_data_0_to_800dys = well_data.query('wlbDrillingDays > 0 and wlbDrillingDays < 800')


plt.figure(figsize=(10, 8))
plt.scatter(well_data_0_to_800dys.wlbTotalDepth, well_data_0_to_800dys.wlbDrillingDays, color='green', edgecolor='k', alpha=0.7)


plt.title('Well Total Depth vs Drilling Days (0 to 800 days)', fontsize=16)
plt.xlabel('Total Depth (ft)', fontsize=14)
plt.ylabel('Drilling Days', fontsize=14)


plt.grid(True)


plt.show()
```

#### 3. Calculate the mean of the total depth according to the "Age At Td", show it as a table, sorted by mean total depth.

```python
mean_depth_by_age = well_data[['wlbAgeAtTd', 'wlbTotalDepth']].groupby('wlbAgeAtTd',).mean().\
    sort_values('wlbTotalDepth', ascending = False)

mean_depth_by_age.rename(columns={'wlbTotalDepth': 'mean_depth'}, inplace=True)

mean_depth_by_age
```

### 4. Production Data :

1. Make a plot with the monthly production data (Oil, Gas and Water) of the `7405` NPDCode Volve according to the time. You will need to add and `monthly` column, create by concatenating the `Year` and the `Month` columns; Plot: vertical axis: the 3 productions (Oil, Gas and Water) and horizontal axis: the month and the year (ex: 2020-01).


2. Calculate the sum of the Oil production by Volve.

```python
monthly_prod.head()
```

```python
monthly_prod.info()
```

```python
well_7405 =  monthly_prod.query('NPDCode == 7405')
well_7405['Date'] = pd.to_datetime(well_7405['Year'].astype(str) + '-' + well_7405['Month'].astype(str))
well_7405.head()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 12), sharex=True)

# Oil data
ax1.plot(well_7405['Date'], well_7405['Oil (Sm3)'], color='orange', marker='o', 
linestyle='-', linewidth=2, markersize=6, label='Oil')
ax1.set_ylabel('Oil (Sm3)', fontsize=14)
ax1.legend(loc='upper left')
ax1.grid(True)

# Gas data
ax2.plot(well_7405['Date'], well_7405['Gas (Sm3)'], color='green', marker='s', 
linestyle='--', linewidth=2, markersize=6, label='Gas')
ax2.set_ylabel('Gas (Sm3)', fontsize=14)
ax2.legend(loc='upper left')
ax2.grid(True)

# Water data
ax3.plot(well_7405['Date'], well_7405['Water (Sm3)'], color='blue', marker='^', 
linestyle='-.', linewidth=2, markersize=6, label='Water')
ax3.set_ylabel('Water (Sm3)', fontsize=14)
ax3.set_xlabel('Date', fontsize=14)
ax3.legend(loc='upper left')
ax3.grid(True)

# Set the title for the entire figure
fig.suptitle('Production Data Over Time', fontsize=18)

# Improve layout
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Display the plot
plt.show()
```

```python
wells_total_productions = monthly_prod[['Wellbore name', 'Oil (Sm3)']].groupby('Wellbore name').sum().\
sort_values('Oil (Sm3)', ascending = False)

wells_total_productions.rename(columns={'Oil (Sm3)': 'total_oil_production'}, inplace=True)


wells_total_productions
```

### Merged Data :

- Create 2 new dataframes:

- partial_wells_data from wells_data by keeping only wlbTotalDepthcolumns and the common column;

- partial_mpd from monthly_production_data (mpd) by keeping only Oil, Gas Water and the common column and making a sum aggregation grouping by the common column;


- Merge both dataframes as merged_data;
- Make a scatter plot of the pil production (vertical axis) according to the total depth (horizontal axis). Is there a correlation between the depth and the production?

```python
print(well_columns_list)
```

```python
partial_wells_data = well_data[['wlbNpdidWellbore', 'wlbTotalDepth']]
```

```python
monthly_prod.columns
```

```python
partial_mpd = monthly_prod[['Oil (Sm3)', 'Gas (Sm3)', 'Water (Sm3)', 'NPDCode' ]]

partial_mpd = partial_mpd.groupby('NPDCode').sum()
```

```python
merged_data = pd.merge(partial_wells_data, partial_mpd, left_on='wlbNpdidWellbore', right_on ='NPDCode')

merged_data
```

```python

plt.figure(figsize=(12, 8))

plt.scatter(merged_data['wlbTotalDepth'], merged_data['Oil (Sm3)'], color='orange', edgecolor='k', alpha=0.7, s=100)

plt.title('Oil Production vs. Well Total Depth', fontsize=16)
plt.xlabel('Well Total Depth (m)', fontsize=14)
plt.ylabel('Oil Production (Sm3)', fontsize=14)


plt.grid(True)


plt.show()
```
